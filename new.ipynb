```python
class RetinalDataset(Dataset):
    def __init__(self, csv_file, image_dir, transform=None):
        self.data = pd.read_csv(csv_file)
        self.image_dir = image_dir
        self.transform = transform
        
        # Ensure column names are consistent
        if 'id_code' not in self.data.columns or 'diagnosis' not in self.data.columns:
            raise ValueError("CSV file must contain 'id_code' and 'diagnosis' columns.")
        
        # Filter out missing files
        self.data = self.data[self.data['id_code'].apply(
            lambda x: os.path.isfile(os.path.join(image_dir, x + '.png'))
        )]

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        img_name = os.path.join(self.image_dir, self.data.iloc[idx, 0] + '.png')
        label = self.data.iloc[idx, 1]
        image = Image.open(img_name).convert("RGB")
        if self.transform:
            image = self.transform(image)
        return image, label
```

```python
import os
from PIL import Image

def convert_images_to_png(image_dir):
    """
    Converts all JPG images in the specified directory to PNG format.
    The converted PNG images will overwrite the original JPG files.
    """
    for img_file in os.listdir(image_dir):
        if img_file.endswith(".jpg") or img_file.endswith(".jpeg"):
            # Input file path
            img_path = os.path.join(image_dir, img_file)
            
            # Open and convert image
            img = Image.open(img_path).convert("RGB")
            
            # Define output file path with .png extension
            output_path = os.path.join(image_dir, os.path.splitext(img_file)[0] + ".png")
            
            # Save image as PNG
            img.save(output_path, "PNG")
            
            # Optionally, remove the original JPG file
            os.remove(img_path)

# Convert train and test images
convert_images_to_png("datasets/EyePACS/train_images")
convert_images_to_png("datasets/EyePACS/test_images")

print("Conversion completed: All JPG images are now PNG.")
```

```python
# Step 3: CNN Backbone
class CNNBackbone(nn.Module):
    def __init__(self):
        super(CNNBackbone, self).__init__()
        self.resnet = resnet50(weights="IMAGENET1K_V1")
        self.resnet.fc = nn.Identity()  # Remove classification head

    def forward(self, x):
        return self.resnet(x)

# Step 4: Graph Neural Network (GNN)
class GCNNetwork(nn.Module):
    def __init__(self, input_dim, hidden_dim, output_dim):
        super(GCNNetwork, self).__init__()
        self.gcn1 = GCNConv(input_dim, hidden_dim)
        self.gcn2 = GCNConv(hidden_dim, output_dim)
        self.norm1 = nn.BatchNorm1d(hidden_dim)
        self.norm2 = nn.BatchNorm1d(output_dim)

    def forward(self, x, edge_index):
        x = self.gcn1(x, edge_index).relu()
        x = self.norm1(x)
        x = self.gcn2(x, edge_index).relu()
        x = self.norm2(x)
        return x


# Step 5: Combined Model Architecture
class GATLModel(nn.Module):
    def __init__(self, num_classes, input_width=64, input_height=64):
        super(GATLModel, self).__init__()
        self.cnn = nn.Sequential(
            nn.Conv2d(3, 64, kernel_size=3, stride=1, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size=2),
            nn.Conv2d(64, 128, kernel_size=3, stride=1, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size=2)
        )
        # Compute flattened feature size
        flattened_size = 128 * (input_width // 4) * (input_height // 4)
        self.fc1 = nn.Linear(flattened_size, num_classes)
        self.fc2 = nn.Linear(flattened_size, num_classes)

    def forward(self, x, edge_index=None):
        cnn_features = self.cnn(x)
        cnn_features = cnn_features.view(cnn_features.size(0), -1)
        if edge_index is not None:
            gnn_features = self.gat(cnn_features, edge_index)
        else:
            gnn_features = cnn_features
        out1 = self.fc1(gnn_features)
        out2 = self.fc2(gnn_features)
        return out1, out2



```

```python
def train(model, source_loader, optimizer, epochs, device):
    criterion = nn.CrossEntropyLoss()
    for epoch in range(epochs):
        model.train()
        total_loss = 0
        print(f"Starting epoch {epoch + 1}/{epochs}...")

        for i, (source_images, source_labels) in enumerate(source_loader):
            print(f"Batch {i + 1}...")
            source_images = source_images.to(device)
            source_labels = source_labels.to(device)

            # Forward pass
            try:
                class_output1, class_output2 = model(source_images, edge_index=None)
                print(f"Batch {i + 1}: Forward pass completed")
            except Exception as e:
                print(f"Error during forward pass: {e}")
                return

            # Compute loss
            loss = criterion(class_output1, source_labels) + criterion(class_output2, source_labels)
            total_loss += loss.item()

            # Backward pass
            optimizer.zero_grad()
            loss.backward()
            optimizer.step()
            print(f"Batch {i + 1}: Loss = {loss.item():.4f}")

        print(f"Epoch [{epoch + 1}/{epochs}], Total Loss: {total_loss:.4f}")
```

```python
from torch.utils.data import Dataset, DataLoader
from PIL import Image
import os

class UnlabeledTestDataset(Dataset):
    def __init__(self, image_dir, transform=None):
        self.image_dir = image_dir
        self.image_files = [f for f in os.listdir(image_dir) if f.endswith(('.png', '.jpg', '.jpeg'))]
        self.transform = transform

    def __len__(self):
        return len(self.image_files)

    def __getitem__(self, idx):
        img_path = os.path.join(self.image_dir, self.image_files[idx])
        image = Image.open(img_path).convert("RGB")
        if self.transform:
            image = self.transform(image)
        return image, self.image_files[idx]  # Return image and filename

# Define your transformations
test_transforms = transforms.Compose([
    transforms.Resize((64, 64)),
    transforms.ToTensor(),
    transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
])

# Create dataset and DataLoader
test_image_dir = "datasets/APTOS/test_images"
test_dataset = UnlabeledTestDataset(test_image_dir, transform=test_transforms)
test_loader = DataLoader(test_dataset, batch_size=32, shuffle=False)

print(f"Loaded {len(test_dataset)} test images.")
```

```python
def evaluate_on_unlabeled_data(model, loader, device):
    model.eval()
    predictions = []

    with torch.no_grad():
        for images, filenames in loader:
            images = images.to(device)
            outputs = model(images)

            # Ensure correct handling of model outputs
            if isinstance(outputs, tuple):
                outputs = outputs[0]  # Extract logits if outputs is a tuple
            
            # Get predicted class
            _, predicted = torch.max(outputs, dim=1)
            predictions.extend(zip(filenames, predicted.cpu().numpy()))

    return predictions
```
